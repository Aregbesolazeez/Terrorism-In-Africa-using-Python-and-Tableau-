#### Definition of terrorism:

noun
1. the unlawful use of violence or threats to intimidate or coerce a civilian population or government, with the goal of furthering political, social, or ideological objectives.
2. the state of fear and submission produced by terrorism.
3. a terrorist method of governing or of resisting a government.
4. intimidation or coercion by instilling fear

Source: Dictionary.com [link](https://www.dictionary.com/browse/terrorism)

#### Obtain Data for Africa Only

```python
# extract data whose region are 10 (Middle East & North Africa) or 10 (Sub-Saharan Africa)
pd.set_option('display.max_rows', 5)
df_africa = df[(df.region == 10) | (df.region==11)]
df_africa.reset_index(inplace = True, drop = True)
df_africa
```

```python
# checks null
pd.set_option('display.max_rows', df_africa.shape[0])
df_africa.isna().sum()
```

```python
# takes mean of column's null values 
df_africa.isnull().mean()
```

```python
#column whose null mean are less than 0.3
df_africa.columns[df_africa.isnull().mean() < 0.3]
```

```python
# remove column whose null mean are greater than 0.3
df_africa = df_africa[df_africa.columns[df_africa.isnull().mean() < 0.3]]
df_africa.head()
```

```python
# gets the numbers of rows and columns in the df_africa
no_rows = df_africa.shape[0]
no_columns = df_africa.shape[1]
f"There are {no_rows} rows and {no_columns} columns in the Africa dataset"
```

#### Regroup Country by Region Northern Africa, Eastern Africa, Middle Africa, Southern Africa and Western Africa According to [Wikipedia](https://en.m.wikipedia.org/wiki/List_of_regions_of_Africa) list.
 Note: Rhodesia is grouped as southern Africa And Zaire as Middle Africa

```python
df_region = pd.read_csv("africaRegion.csv", encoding="ISO-8859-1")
df_region
```

#### Join datasets

```python
pd.set_option('display.max_rows', 5)
df = df_africa.merge(df_region, on='country_txt', how='left')
df
```

#### Ensure proper joining

```python
# consider countries in Sub-Saharan Africa
df2 = df[df['region_txt'] == 'Sub-Saharan Africa']
# check countries that do not have a subregion
df2[df2.Subregion.isna()].country_txt.unique()
```

```python
# consider countries in Middle East & North Africa
df3 = df[df['region_txt'] == 'Middle East & North Africa']
# countries that belong to North Africa
df3[df3.Subregion.notna()].country_txt.unique()
```

```python
# countries that belong to Middle East
df3[df3.Subregion.isna()].country_txt.unique()
```

```python
# drop country that belongs to Middle East in df
pd.set_option('display.max_columns', None)
df = df[df.Subregion.notna()]
df.head()
```
